```python
def get_klines(self, symbol, start_time, end_time, interval="1m"):
    url = f"https://openapi.bitrue.com/api/v1/market/klines?symbol={symbol}&interval={interval}&startTime={start_time}&endTime={end_time}"
    response = requests.get(url)
    print(response.json())
    return response.json()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('orders_profit_5h.csv')
data_btc = data[data['symbol'] == 'BTCUSDT']
data_eth = data[data['symbol'] == 'ETHUSDT']
data_xrp = data[data['symbol'] == 'XRPUSDT']

results = pd.DataFrame(columns=['Symbol', 'Total Profit', 'Total Loss', 'Max Profit', 'Max Loss', 'Mean Profit', 'Mean Loss'])

for symbol, symbol_data in [('BTC', data_btc), ('ETH', data_eth), ('XRP', data_xrp)]:
    total_profit = int(symbol_data['max_profit'].sum())
    total_loss = int(symbol_data['max_loss'].sum())
    max_profit = int(symbol_data['max_profit'].max())
    max_loss = int(symbol_data['max_loss'].max())
    mean_profit = int(symbol_data['max_profit'].mean())
    mean_loss = int(symbol_data['max_loss'].mean())


    results = results.append({
        'Symbol': symbol,
        'Total Profit': total_profit,
        'Total Loss': total_loss,
        'Max Profit': max_profit,
        'Max Loss': max_loss,
        'Mean Profit': mean_profit,
        'Mean Loss': mean_loss
    }, ignore_index=True)

results.set_index('Symbol', inplace=True)

# 绘制表格
fig, ax = plt.subplots(figsize=(12, 3))  # 调整表格大小
ax.axis('off')  # 隐藏坐标轴

table = ax.table(cellText=results.values,
                 colLabels=results.columns,
                 rowLabels=results.index,
                 cellLoc='center',
                 loc='center')

table.auto_set_font_size(False)  # 禁用自动设置字体大小
table.set_fontsize(14)  # 设置字体大小
table.scale(1, 1.5)  # 调整表格行高

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('orders_profit_5h.csv')

def plot_data(symbol, data):
    symbol_data = data[data['symbol'] == symbol]
    
    plt.figure(figsize=(12, 6))
    plt.plot(symbol_data['id'], symbol_data['price'], color='black', linewidth=1.5, label='Price')
    plt.plot(symbol_data['id'], symbol_data['high'], color='green', linewidth=1.5, linestyle='--', label='High')
    plt.plot(symbol_data['id'], symbol_data['low'], color='red', linewidth=1.5, linestyle='--', label='Low')

    plt.legend()
    plt.xlabel('ID')
    plt.ylabel('Value')
    plt.title(f'{symbol} Price, High and Low')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
plot_data('XRPUSDT', data)
plot_data('BTCUSDT', data)
plot_data('ETHUSDT', data)
```
